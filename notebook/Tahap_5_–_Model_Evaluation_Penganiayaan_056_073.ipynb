```python
from google.colab import drive
drive.mount('/content/drive')

import matplotlib.pyplot as plt
import pandas as pd
import os

# === Simulasi variabel metrik model utama ===
model_names = ['TF-IDF', 'BERT', 'SVM']
accuracy = [0.3, 0.6, 0.7]
precision = [0.4, 0.7, 0.75]
recall = [0.35, 0.65, 0.72]
f1 = [0.36, 0.67, 0.73]

df_metrics = pd.DataFrame({
    'Model': model_names,
    'Accuracy': accuracy,
    'Precision': precision,
    'Recall': recall,
    'F1-score': f1
})

# === Simulasi isi retrieval_metrics_df dan prediction_metrics_dict ===
retrieval_metrics_df = pd.DataFrame({
    'query_id': ['query_001', 'query_002', 'query_003'],
    'ground_truth': ['case_001', 'case_002', 'case_003'],
    'top_k_hits': [['case_005','case_001'], ['case_002'], ['case_010']],
    'hit_at_1': [0, 1, 0],
    'hit_at_k': [1, 1, 1],
})

prediction_metrics_dict = {
    'accuracy': 0.33,
    'precision_weighted': 0.40,
    'recall_weighted': 0.33,
    'f1_weighted': 0.30
}

# === Direktori penyimpanan ===
eval_dir = "/content/drive/MyDrive/SEMESTER 6/Penalaran Komputer/UAS_Penalaran Komputer/data/eval"
os.makedirs(eval_dir, exist_ok=True)

# === Simpan retrieval_metrics.csv ===
retrieval_path = os.path.join(eval_dir, "retrieval_metrics.csv")
retrieval_metrics_df.to_csv(retrieval_path, index=False)
print(f"‚úîÔ∏è retrieval_metrics.csv disimpan di: {retrieval_path}")

# === Simpan prediction_metrics.csv ===
prediction_path = os.path.join(eval_dir, "prediction_metrics.csv")
pd.DataFrame([prediction_metrics_dict]).to_csv(prediction_path, index=False)
print(f"‚úîÔ∏è prediction_metrics.csv disimpan di: {prediction_path}")

# === Buat dan Simpan Bar Chart ===
plt.figure(figsize=(10, 6))
df_metrics.set_index('Model')[['Accuracy', 'Precision', 'Recall', 'F1-score']].plot(
    kind='bar', ylim=(0, 1), colormap='Set2', rot=0, figsize=(10, 6)
)
plt.title("Perbandingan Kinerja Retrieval dan Klasifikasi")
plt.ylabel("Nilai Skor")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()

# Simpan chart ke PNG
chart_path = os.path.join(eval_dir, "model_performance_chart.png")
plt.savefig(chart_path)
plt.show()
print(f"üìä Chart disimpan di: {chart_path}")
```
